```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
athletes = pd.read_csv("D:/Study/Data Science Content/Datasets/2020 Olympics Data Analysis/athlete_events.csv")
regions = pd.read_csv("D:/Study/Data Science Content/Datasets/2020 Olympics Data Analysis/noc_regions.csv")
```

```python
regions.head()
```

```python
# Join the Dataframes

athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df
```

```python
# Column name consistent
athletes_df.rename(columns = {'region' : "Region", 'notes' : 'Notes'}, inplace = True)
```

```python
# Top Countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries
plt.figure(figsize = (12, 6))
plt.xticks(rotation = 20)
plt.title('Overall Participation by Country')
sns.barplot(x = top_10_countries.index, y = top_10_countries, palette = 'Set2');
```

```python
# Age Disctibution of the Participants
plt.figure(figsize = (12, 6))
plt.title("Age Distribution of the athletes")
plt.xlabel('Age')
plt.ylabel("Number of Participants")
plt.hist(athletes_df.Age, bins = np.arange(10, 80, 2), color = 'orange', edgecolor = 'white');
```

```python
# Male and Female Participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for male and female athletes (autopct variable enable to display the percent value usign python)
plt.figure(figsize = (12, 6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle=150, shadow = True);
```

```python
# Total number of female athletes in each olympics.

female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize = (20, 10))
sns.countplot(x = 'Year', data = womenOlympics, palette = "Spectral")
plt.title('Women Participation')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# Gold medal indian athletes 
goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals
goldmedals_for_india = athletes_df[(athletes_df.Medal == 'Gold') & (athletes_df.Team == 'India')]
goldmedals_for_india.head(10)
```

```python
# Take only the values that are different from NaN.

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Gold beyond age of 60 years
goldMedals['ID'][goldMedals['Age'] >= 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] >= 60]
sporting_event
```

```python
# Plot for sporting event
plt.figure(figsize=(10, 5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medals for Athletes over or equal to 60 years of age.")
```

```python
# Gold medals for each country
goldMedals_world = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals_world.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals = goldMedals_world.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index', y = 'Medal', data = totalGoldMedals, height=5, kind="bar", palette= "rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
```

```python
# Rio Olympic

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == "Gold")].Team

team_names.value_counts().head(10)
```

```python
sns.barplot(x = team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016')
```
